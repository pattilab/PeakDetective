```python
import os
import sys
sys.path.append("../src/")
import importlib
import integrAitor
importlib.reload(integrAitor)
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
```

```python
processor = integrAitor.rawData()
```

```python
data = pd.read_csv("../data/formatted_data.csv")
uniqueFiles = list(set(data["sample"].values))
print(len(uniqueFiles))
data
```

```python
metInfo = pd.read_csv("../data/metabolite_info_negative.csv",index_col=0)
metInfo
```

```python
ppm = 40
true = []
pred = []
for file in uniqueFiles:
    filt = data[data["sample"] == file]
    processor.readRawDataFile(filt.at[filt.index.values[0],"path"],ppm)
    for index,row in filt.iterrows():
        mz = metInfo.at[row["metabolite"],"Precursor Mz"]
        rt_start = row["rt_start"]
        rt_end = row["rt_end"]
        tmp = pd.DataFrame.from_dict({0:{"mz":mz,
                                         "rt_start":rt_start,
                                         "rt_end":rt_end}},orient="index")
        true.append(np.log2(row["area"]))
        pred.append(np.log2(processor.integrateTargets(tmp)[row["path"]].values[0]))
        
        #rts,intensity = processor.extractEIC(mz,rt_start,rt_end,ppm)
        #plt.plot(rts,intensity)
        #plt.title(row["metabolite"])
        #plt.figure()
    #break
plt.scatter(true,pred,s=.1)
```

```python
plt.scatter(true,pred,s=0.1)
```

```python
bad = []
errs = []
for x in range(len(true)):
    err = np.abs(true[x] - pred[x])
    if not np.isinf(err):
        errs.append(true[x]-pred[x])
        if np.abs(true[x]-pred[x]) > 10 or np.abs(true[x]-pred[x]) < 5:
            bad.append(x)
print(len(bad),len(true))
plt.hist(errs,bins=20)
```

```python
plt.scatter([true[x] for x in range(len(true)) if x not in bad],[pred[x] for x in range(len(true)) if x not in bad],s=0.1)
```

```python
ind = 0
goodInds = []
for file in uniqueFiles:
    filt = data[data["sample"] == file]
    for index,row in filt.iterrows():
        if ind not in bad:
            goodInds.append(index)
print(len(goodInds))
data_filt = data.loc[goodInds,:]
```
