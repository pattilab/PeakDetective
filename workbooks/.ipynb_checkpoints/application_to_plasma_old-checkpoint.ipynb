```python

files = [x for x in os.listdir(datadir) if ".mzML" in x]

print(files)

ppm = 10

raw_data = {}

for file in files:
    data = deimos.read_mzml(datadir + file)
    deimos.save_hdf(datadir + file.replace(".mzML",'.h5'), data['ms1'], key='ms1', mode='w')
    
    temp = rawData()
    temp.readRawDataFile(datadir + file,ppm)
    
    raw_data[file] = temp
    
```

```python
if __name__ == "__main__":

    # initialize ms1 peak container
    ms1_peaks = {}

    # iterate over ms1 datasets
    for k in ms1.keys():
        # partition data into chunks to limit memory usage
        # more on partitioning/mapping later in this notebook
        partitions = deimos.partition(ms1[k],
                                      split_on='mz',
                                      size=500,
                                      overlap=0.05)

        # map peakpicking algorithm to each partition
        # note that more configuration parameters are available
        ms1_peaks[k] = partitions.map(deimos.peakpick.local_maxima,
                                      dims=['mz', 'retention_time'],bins=[2.7, 3.64],
                                      processes=10)

        # save result
        deimos.save_hdf(datadir + k.replace(".mzML",'_peaks.h5'), ms1_peaks[k], key='ms1', mode='w')
```

```python
#format X
integ = integrAitor(numCores = 10)
if __name__ == "__main__":
    X = np.zeros((int(np.sum([len(ms1_peaks_thresh[f]) for f in files])),integ.resolution))
    i = 0
    keys = []

    for file in files:
        print(file)
        tmp = integ.makeDataMatrix([raw_data[file]],ms1_peaks_thresh[file]["mz"].values,
                                 ms1_peaks_thresh[file]["retention_time"].values-.5,
                                 ms1_peaks_thresh[file]["retention_time"].values+.5)
        keys += [[file,index] for index in ms1_peaks_thresh[file].index.values]
        X[i:i+len(tmp),:] = tmp        
        i += len(tmp)

X = X[:i]
print(X.shape)
```

```python
def safeNormalize(x):
    if np.sum(x) < 1e-6:
        tmp = np.ones(x.shape) 
        return tmp / np.sum(tmp)
    else:
        return x/np.sum(x)
```

```python
X_norm = np.array([safeNormalize(x) for x in X])
tics = np.log2(np.array([np.max([2,np.sum(x)]) for x in X]))
```

```python
pca = PCA()
loads = pca.fit_transform(X_norm)
pca.explained_variance_ratio_[:2]
```

```python
plt.scatter(loads[:,0],loads[:,1],c=tics,s=.1,alpha=.5)
plt.xlabel("PC1 (" + str(np.round(100*pca.explained_variance_ratio_[0],2)) + "%)")
plt.ylabel("PC2 (" + str(np.round(100*pca.explained_variance_ratio_[1],2)) + "%)")
plt.colorbar(label="log2(intensity)")
```

```python
#build autoencoder
autoencoderInput = keras.Input(shape=(integ.resolution,))
#x = layers.Softmax()(autoencoderInput)
x = layers.Reshape((100,1))(autoencoderInput)

kernelsize=3
stride=1
max_norm_value = 2.0

x = layers.Conv1D(64, kernelsize, strides=stride, activation='relu',kernel_constraint=max_norm(max_norm_value), kernel_initializer='he_uniform')(x)

x = layers.Conv1D(16, kernelsize, strides=stride, activation='relu',kernel_constraint=max_norm(max_norm_value), kernel_initializer='he_uniform')(x)

#x = layers.Conv1DTranspose(16, kernelsize, strides=stride, activation='relu',kernel_constraint=max_norm(max_norm_value), kernel_initializer='he_uniform')(x)

x = layers.Flatten()(x)

#x = layers.Dense(100,activation="relu")(x)

x = layers.Dense(20,activation="relu")(x)

#x = layers.Dense(100,activation="relu")(x)

x = layers.Dense(1536,activation="relu")(x)

x = layers.Reshape((96,16))(x)

x = layers.Conv1DTranspose(64, kernelsize, strides=stride, activation='relu',kernel_constraint=max_norm(max_norm_value), kernel_initializer='he_uniform')(x)

#x = layers.Conv1DTranspose(1, kernelsize, strides=stride, activation='relu',kernel_constraint=max_norm(max_norm_value), kernel_initializer='he_uniform')(x)
x = layers.Conv1DTranspose(1, kernelsize, strides=stride, activation='sigmoid',kernel_constraint=max_norm(max_norm_value), kernel_initializer='he_uniform')(x)


#x = layers.Conv1D(1, kernelsize, strides=stride, padding='same',activation="sigmoid",kernel_constraint=max_norm(max_norm_value))(x)

x = layers.Flatten()(x)

#x = layers.Dense(100,activation="sigmoid")(x)

#x = layers.Softmax()(x)


autoencoder = keras.Model(autoencoderInput,x)

autoencoder.compile(loss='binary_crossentropy', optimizer=keras.optimizers.Adam(1e-4), metrics=['mean_absolute_error'])

autoencoder.save_weights('autoencoder.h5')

autoencoder.summary()
```

```python
if __name__ == "__main__":
    autoencoder.load_weights('autoencoder.h5')
    autoencoder.fit(X_norm,X_norm,epochs=10,batch_size=64,validation_split=.1)
```

```python
if __name__ == "__main__":
    
    X_noise = np.array(X_norm).flatten()
    np.random.shuffle(X_noise)
    X_noise = np.reshape(X_noise,X_norm.shape)
    X_noise = X_noise / np.sum(X_noise,axis=1)[:,np.newaxis]
    noise_tics = np.array(tics)
    np.random.shuffle(noise_tics)
    X_noise_smoothed = autoencoder.predict(X_noise)
```

```python
if __name__ == "__main__":
    tmp = np.array(X_norm).flatten()
    np.random.shuffle(tmp)
    tmp = tmp.reshape((X_norm.shape))
    tmp = tmp / tmp.sum(axis=1)[:,np.newaxis]
    X_signal = np.array([[stats.norm.pdf(x,.5,.1*s+.001) for x in np.linspace(0,1,100)] for x,s in zip(range(len(X_noise)),np.random.random(len(X_noise)))])
    X_signal = X_signal / np.sum(X_signal,axis=1)[:,np.newaxis]
    s2n = .5 + .3*np.random.random(len(X_signal))
    X_signal = s2n[:,np.newaxis] * X_signal + (1-s2n[:,np.newaxis]) * tmp
    signal_tics = np.array(tics)
    np.random.shuffle(signal_tics)
    X_signal_smoothed = autoencoder.predict(X_signal)
```

```python
y = np.array([[.5,.5] for _ in X_smoothed] + [[1,0] for _ in X_noise_smoothed] + [[0,1] for _ in X_signal_smoothed])
X_merge = np.concatenate((X_smoothed,X_noise_smoothed,X_signal_smoothed),axis=0)
tic_merge = np.concatenate((tics, noise_tics,signal_tics))
tic_merge = np.ones(tic_merge.shape)
print(y.shape,X_merge.shape,tic_merge.shape)
```

```python
loads = pca.fit_transform(X_merge)
print(pca.explained_variance_ratio_[:2])
def getColor(s):
    if s < 0.01:
        return "black"
    if s > 0.99:
        return "red"
    else:
        return "grey"
cs = [getColor(x[1]) for x in y]
plt.scatter([0],[0],color="black",label="simulated noise",s=1,alpha=.5)
plt.scatter([0],[0],color="red",label="simulated guassians",s=1,alpha=.5)
plt.scatter([0],[0],color="grey",label="measured peaks",s=1,alpha=.5)

plt.scatter(loads[:,0],loads[:,1],s=.1,alpha=.5,c=cs)
plt.xlabel("PC1 (" + str(np.round(100*pca.explained_variance_ratio_[0],2)) + "%)")
plt.ylabel("PC2 (" + str(np.round(100*pca.explained_variance_ratio_[1],2)) + "%)")
plt.legend()
```

```python
descriminatorInput = keras.Input(shape=(integ.resolution,))
ticInput = keras.Input(shape=(1,))

kernelsize=3
stride=2

x = layers.Reshape((100,1))(descriminatorInput)
x = layers.Conv1D(8, kernelsize, strides=stride, activation='relu',kernel_constraint=max_norm(max_norm_value), kernel_initializer='he_uniform')(x)

#x = layers.Conv1D(32, kernelsize, strides=stride, activation='relu',kernel_constraint=max_norm(max_norm_value), kernel_initializer='he_uniform')(x)

#x = layers.Conv1D(16, kernelsize, strides=stride, activation='relu',kernel_constraint=max_norm(max_norm_value), kernel_initializer='he_uniform')(x)

x = layers.Conv1D(4, kernelsize, strides=stride, activation='relu',kernel_constraint=max_norm(max_norm_value), kernel_initializer='he_uniform')(x)

x = layers.Flatten()(x)

x = layers.Dense(20, activation="relu")(x)

x = keras.Model(descriminatorInput,x)

tic = keras.Model(ticInput,layers.Dense(1,activation="linear")(ticInput))

x = layers.concatenate([x.output,tic.output],axis=1)
x = layers.Dense(10, activation="relu")(x)
output = layers.Dense(2, activation="softmax")(x)

discriminator = keras.Model([descriminatorInput,ticInput], output, name="discriminator")

discriminator.summary()

discriminator.compile(loss='binary_crossentropy', optimizer=keras.optimizers.Adam(1e-4), metrics=['mean_absolute_error'])

discriminator.save_weights('model.h5')
```

```python
def classify(preds):
    classes = np.zeros(preds.shape)
    for x in range(len(preds)):
        classes[x,list(preds[x]).index(np.max(list(preds[x])))] = 1
    return classes
```

```python
#for smoothed
if __name__ == "__main__":
    synInds = [x for x in range(len(y)) if y[x][1] < .25 or y[x][1] > .75]
    realInds = [x for x in range(len(y)) if y[x][1] > .25 and y[x][1] < .75]
    updatingInds = list(realInds)
    trainingInds = list(synInds)
    numRemaining = []
    for i in range(10):
        print(len(updatingInds))
        numRemaining.append(len(updatingInds))
        discriminator.load_weights('model.h5')
        discriminator.fit([X_merge[trainingInds],tic_merge[trainingInds]],y[trainingInds],epochs=10,batch_size=64,validation_split=.1)
        scores = discriminator.predict([X_merge[updatingInds],tic_merge[updatingInds]])
        for ind,s in zip(updatingInds,scores):
            if s[1] < 0.05 or s[1] > 0.95:
                y[ind] = classify(s.reshape(1,-1))[0]
        trainingInds = [x for x in range(len(y)) if y[x][1] < .25 or y[x][1] > .75]
        updatingInds = [x for x in range(len(y)) if y[x][1] > .25 and y[x][1] < .75]
    
    
scores = discriminator.predict([X_merge[updatingInds],tic_merge[updatingInds]])
y[updatingInds] = classify(scores)
discriminator.load_weights('model.h5')
discriminator.fit([X_merge,tic_merge],y,epochs=10,batch_size=64,validation_split=.1)

```

```python
ms1_peaks_curated = {file:[] for file in files}
for [file,index],score in zip(keys,y[realInds][:,1]):
    if score > .5:
        ms1_peaks_curated[file].append(index)
for file in ms1_peaks_curated:
    ms1_peaks_curated[file] = ms1_peaks_thresh[file].loc[ms1_peaks_curated[file],:]
```

```python
df = {}
for file in ms1_peaks_thresh:
    for index,row in ms1_peaks_thresh[file].iterrows():
        if index in ms1_peaks_curated[file].index.values:
            df[len(df)] = {"sample":file,"classification":"true peak"}
        else:
            df[len(df)] = {"sample":file,"classification":"artifact"}

df = pd.DataFrame.from_dict(df,orient="index")

def aggregate(rows,columns,df):
    column_keys = df[columns].unique()
    row_keys = df[rows].unique()

    agg = { key : [ 100*len(df[(df[rows]==value) & (df[columns]==key)]) / len(df[df[rows]==value]) for value in row_keys]
               for key in column_keys }

    aggdf = pd.DataFrame(agg,index = row_keys)
    aggdf.index.rename(rows,inplace=True)

    return aggdf

aggregate('sample','classification',df).plot(kind='bar',stacked=True,color=["blue","grey","black"])
plt.xticks([],[])
plt.ylabel("% of features ")
```
